### We use requests to grab html from page, beauitfulsoup for parsing the html and re to write some regex to find what we are looking for

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.transfermarkt.com/premier-league/transfers/wettbewerb/GB1/plus/?saison_id=2017&s_w=&leihe=0&intern=0' 
page = requests.get(url , headers={'User-Agent': 'Custom'})

# code of 200 means request was sucessful
page.status_code 
```

```python
# Prase the raw html from our request using beautifulsoup
soup = BeautifulSoup(page.content, 'lxml')
```

```python
# The transfer fee info we are looking for comes under 2 class tags
# We write a regex that is passed to find_all() to get them both
regex = re.compile('^rechts $|^rechts bg_blau_20$')

# Use find_all() to get the html info/tags we are looking for
fees = soup.find_all('td',  {'class' : regex})
players  = soup.find_all('span', class_='hide-for-small')
```

```python
list_players = []
list_fees = []

# Use regex to remove all html tags leaving just the string at the centre
for player in players:
    name = str(re.sub("<.*?>", "", str(player)))
    list_players.append(name)

for fee in fees:
    price = str(re.sub("<.*?>", "", str(fee)))
    list_fees.append(price)
```

```python
list_players.remove('') # There is an empty entry we need to remove
```

```python
print(len(list_players))
print(len(list_fees))
```

```python
player_fee_combined = zip(list_players, list_fees)
```
